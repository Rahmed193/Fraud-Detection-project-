**Fraud Rate**

```python
fraud_rate = df["is_fraud"].mean()*100
print(f"Fraud Rate: {fraud_rate:.2f}%")
```

**High-Risk Country Analysis**

```python
country_fraud = (
    df.groupby("country").agg(total_transactions=("transaction_id", "count"),
                              fraud_cases = ("is_fraud", "sum")
    )

)

country_fraud["fraud_rate"]= (country_fraud["fraud_cases"]/ country_fraud["total_transactions"]) *100
country_fraud.sort_values(by="fraud_rate", ascending=False)
```

**Fraud by Merchant Category**

```python
merchant_fraud = (
    df.groupby("merchant_category").agg(
        fraud_cases = ("is_fraud", "sum"),
        total_transactions = ("transaction_id", "count")
    )
)

merchant_fraud["fraud_rate"] = (merchant_fraud["fraud_cases"]/ merchant_fraud["total_transactions"]) *100

merchant_fraud.sort_values(by="fraud_rate", ascending=False)
```

**Fraud by hour**

```python
hourly_fraud = (df.groupby ("hour").agg(
    fraud_cases = ("is_fraud", "sum"),
    total= ("transaction_id", "count")
)
)

hourly_fraud["fraud_rate"] = (hourly_fraud["fraud_cases"]/ hourly_fraud["total"]) *100

hourly_fraud.sort_values("fraud_rate", ascending=False)
```
